# BeautifulSoup 코드 분석
> 1. html 분석하기
    - soup = BeautifulSoup(html,'html.parser')
> 1. find() 메서드를 이용한 데이터 훔치기: id 속성 값으로 data를 추출
    - title =soup.find(id="title")
    - subTitle = soup.find(id="subTitle")

```python
from bs4 import BeautifulSoup
html="""
<html><body>
    <h1 id="title">BeautifulSoup</h1>
    <p id="subTitle">스크레이핑 연습하기</p>
    <P>원하는 데이터 추출</p>
</body></html>
"""
print(html)
```

```python
soup = BeautifulSoup(html,'html.parser')
soup
```

```python
title = soup.find(id="title")
print(title)
print(title.string)
```

```python
subTitle = soup.find(id="subTitle")
print(subTitle)
print(subTitle.string)
```

> BeautifulSoup
    - find_all()사용하기

```python
from bs4 import BeautifulSoup
html = """
<html>
    <body>
        <ul>
            <li><a href="http://www.naver.com">네이버</a></li>
            <li><a href="http://www.daum.net">다음</a></li>
        </ul>
    </body>
</html>    
"""
soup = BeautifulSoup(html,'html.parser')
print(soup)
```

```python
print(html)
```

```python
a1 = soup.html.body.ul.li.a
print(a1)
```

```python
print(a1.string)
```

```python
links = soup.find_all("a")
links

# 리스트 값처럼 되어있는 것을 알 수 있다
```

```python
print(type(links))
```

```python
links[0]
print("links[0]:",links[0])
```

```python
# a가 나오는 h태그 본문을 가져오자!

for a in links:
    print(a.string)
```

```python
# www.naver.com 링크주소를 추출해보자!
for a in links:
    href = a.attrs['href']
    text = a.string
    print(text,">",href)
    
    
# attrs : atrributes 의 약자!!!!
```

# 크롤링 실습
## 1.BeautifulSoup은 Data Download 기능이 없다
### 따라, urllib.request를 사용해서 다운로드를 가져와라
> 1. 분석기 종류 선택 beautifulSoup
> 1. 다운로드 기능을 포함하는 url라이브러리를 request 하자
> 1. url 사이트를 불러오자
> 1. response 값의 응답을 받아오자. =req.urlopen() > 파일이 끝나기전에 저장

```python
from bs4 import BeautifulSoup
import urllib.request as req

url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"

res = req.urlopen(url)

soup = BeautifulSoup(res, "html.parser")

# soup 변수에다가 response값을 저장해라

print(soup)
```

## 2. 요소와 그 데이터 접근

```python
title = soup.find("title")
print(title.string)
```

```python
wf = soup.find("wf")
print(wf.string)
```
